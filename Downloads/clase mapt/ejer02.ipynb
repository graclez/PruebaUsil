GRAFICO DE DISPERSION

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python

df_ventas = pd.DataFrame({
    'inversion_publicidad': [100, 150, 200, 250, 300],
    'ventas_miles': [250, 300, 380, 450, 510]
})

sns.scatterplot(x='inversion_publicidad', y='ventas_miles', data=df_ventas, color ='green'  )
plt.title('Relación entre Inversión y Ventas')
plt.xlabel('Inversión en Publicidad')
plt.ylabel('Ventas (en miles)')
plt.show()
```

```python
df_ventas = pd.DataFrame({
    'inversion_publicidad': [100, 150, 200, 250, 300],
    'ventas_miles': [250, 300, 200, 450, 510]
})

sns.regplot(
    x='inversion_publicidad',
    y='ventas_miles',
    data=df_ventas,
    color='darkorange',     
    marker='o' 
)

plt.title('Relación entre Inversión y Ventas (con tendencia)')
plt.xlabel('Inversión en Publicidad')
plt.ylabel('Ventas (en miles)')
plt.show()
```

CORRELACION

```python
df = pd.read_excel("Financial Sample.xlsx")
df.columns = df.columns.str.strip()

plt.figure(figsize=(7,4))
sns.scatterplot(data=df, x="Sales", y="Profit")
plt.title("Ventas vs Ganancia (dispersión simple)")
plt.xlabel("Sales ($)")
plt.ylabel("Profit ($)")
plt.tight_layout()
plt.show()
```

CAJAS

```python


df_casas = pd.DataFrame({
    'barrio': ['A', 'A', 'B', 'B', 'C', 'C'],
    'precio': [250, 280, 210, 230, 300, 320]
})

sns.boxplot(x='barrio', y='precio', data=df_casas,    palette='Set2'  )
plt.title('Distribución de Precios por Barrio')
plt.xlabel('Barrio')
plt.ylabel('Precio (en miles)')
plt.show()
```

```python
df = pd.read_excel("Financial Sample.xlsx")

df.columns = df.columns.str.strip()

plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Segment", y="Profit", color="Orange")

plt.title("Distribución de Ganancia por Segmento")
plt.xlabel("Segmento de Cliente")
plt.ylabel("Ganancia")
plt.tight_layout()
plt.show()
```

Violin


```python

sns.violinplot(x='barrio', y='precio', data=df_casas)
plt.title('Densidad y Distribución de Precios por Barrio')
plt.show()
```

HUE

```python

df_ventas = pd.DataFrame({
    'inversion_publicidad': [100, 150, 200, 250, 300],
    'ventas_miles': [250, 300, 380, 450, 510]
})


df_ventas['tipo_mercado'] = ['Online', 'Offline', 'Online', 'Offline', 'Online']


sns.scatterplot(
    x='inversion_publicidad',
    y='ventas_miles',
    hue='tipo_mercado',
    data=df_ventas
)

plt.title('Inversión vs. Ventas por Tipo de Mercado')
plt.xlabel('Inversión en Publicidad')
plt.ylabel('Ventas (en miles)')
plt.show()
```

EJERCICIO FINAL

```python
df = pd.read_excel('Financial Sample.xlsx')
df.columns = df.columns.str.strip()
```

para ver la relación entre Sales (ventas) y Profit (ganancia)

```python
sns.scatterplot(data=df, x='Sales', y='Profit')
plt.title('Relación entre Ventas y Ganancia')
plt.show()
```

para ver si hay diferencias entre segmentos de clientes (Segment):

```python
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Segment', palette='Set2')
plt.title('Ventas vs Ganancia por Segmento')
plt.show()
```

para comparar la distribución de ganancias (Profit) según el país (Country):

```python
sns.boxplot(data=df, x='Country', y='Profit', palette='Set3')
plt.title('Distribución de Ganancia por País')
plt.xticks(rotation=45)
plt.show()
```
