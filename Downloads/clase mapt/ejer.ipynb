```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer y limpiar
df = pd.read_excel("Financial Sample.xlsx")
df.columns = df.columns.str.strip()

# Agrupar por país y sumar las ventas
ventas_pais = df.groupby("Country")["Sales"].sum()

# Lista de colores personalizados (uno por país)
colores = [ "#2ca02c"]  # azul, naranja, verde, violeta, rojo


# Graficar
#plt.bar(ventas_pais.index, ventas_pais.values)
plt.bar(ventas_pais.index, ventas_pais.values, color=colores)
plt.title("Ventas totales por país")
plt.xlabel("País")
plt.ylabel("Ventas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

comparar ganancias
cogs: costo de los bienes vendidos - profit: ganacia


```python
profit_cost = df.groupby("Country")[["Profit", "COGS"]].sum()

profit_cost.plot(kind="bar", figsize=(8,5))
plt.title("Ganancia vs Costo por País")
plt.xlabel("País")
plt.ylabel("Monto")
plt.xticks(rotation=45)
plt.show()
```

VENTAS POR MES

```python
df["Date"] = pd.to_datetime(df["Date"])
ventas_mes = df.groupby(df["Date"].dt.to_period("M"))["Sales"].sum()

ventas_mes.plot(kind="line",  color ="green" ,markerfacecolor="orange", markeredgecolor="purple" ,marker="o", figsize=(8,5))
plt.title("Evolución de Ventas Mensuales")
plt.xlabel("Mes")
plt.ylabel("Ventas")
plt.grid(True)
plt.show()
```

Promedio de Ganancia por producto


ascending = False

```python
promedio_profit = df.groupby("Product")["Profit"].mean().sort_values()

promedio_profit.plot(kind="barh", color="red", figsize=(8,5))
plt.title("Promedio de Ganancia por Producto")
plt.xlabel("Ganancia Promedio")
plt.ylabel("Producto")
plt.show()
```

```python
# 2️⃣ Crear los datos
alumnos = ['Luisa', 'Marco', 'Elena']
notas_t1 = [75, 88, 92]
notas_t2 = [80, 90, 89]

# -------------------------------------------------
# PARTE 1: GRÁFICO DE BARRAS (Notas del Trimestre 1)
# -------------------------------------------------
plt.figure(figsize=(6,4))  # tamaño del gráfico (ancho x alto)

# Crear las barras
plt.bar(alumnos, notas_t1, color='skyblue')

# Título y etiquetas
plt.title("Notas del Trimestre 1", fontsize=13)
plt.xlabel("Alumnos")
plt.ylabel("Notas")

# Ajustar el eje Y para mostrar de 0 a 100
plt.ylim(0, 100)

# Mostrar el gráfico
plt.show()

# -------------------------------------------------
# PARTE 2: GRÁFICO DE LÍNEAS (Comparación T1 vs T2)
# -------------------------------------------------
plt.figure(figsize=(7,4))  # nuevo gráfico

# Dibujar la línea del Trimestre 1
plt.plot(alumnos, notas_t1, marker='o', color='blue', label="Trimestre 1")

# Dibujar la línea del Trimestre 2
plt.plot(alumnos, notas_t2, marker='s', color='green', label="Trimestre 2")

# Títulos y etiquetas
plt.title("Comparación de Notas: T1 vs T2", fontsize=13)
plt.xlabel("Alumnos")
plt.ylabel("Notas")

# Mostrar leyenda y cuadrícula
plt.legend()      # muestra los nombres de las líneas
plt.grid(True)    # activa la cuadrícula para ver tendencias

# Limitar el eje Y de 0 a 100
#plt.ylim(0, 100)

# Mostrar el gráfico
plt.show()
```
